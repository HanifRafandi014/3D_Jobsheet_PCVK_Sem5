# Transformasi Linier Brightness

import library data

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

Mengubah tingkat brightness citra

```python
print(' Mengubah tingkat kecerahan citra ')
print('----------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/File PCVK/shark.jpeg')
brightness_image = np.zeros(original.shape, original.dtype)

# akses per pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)

# cara simple tanpa for loop
# brightness_image = cv.convertScaleAbs(original, beta-brightness)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

2. Implementasikan transformasi contrast pada Google Colaboratory menggunakan
formula yang terdapat pada bagian Ulasan Teori untuk kontras, sehingga
menghasilkan keluaran seperti berikut:

```python
print(' Mengubah kontras dan tingkat kecerahan citra ')
print('----------------------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan [-255 - 255]: '))
except ValueError:
  print('Error, not a number')

try:
  contrast = int(input('Masukkan kontras [1.0 - 3.0]: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/File PCVK/shark.jpeg')
brightness_image = np.zeros(original.shape, original.dtype)

# akses per pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(contrast * original[y,x,c] + brightness, 0, 255).astype(np.uint8)

# cara simple tanpa for loop
# brightness_image = cv.convertScaleAbs(original, beta-brightness)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

3. Implementasikan transformasi logarithmic brightness pada Google Colaboratory
menggunakan formula yang terdapat pada bagian Ulasan Teori untuk transformasi
log, sehingga menghasilkan keluaran seperti berikut:

```python
print(' Mengubah kontras dan tingkat kecerahan citra dengan Transformasi Log ')
print('----------------------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/File PCVK/shark.jpeg')
brightness_image = np.zeros(original.shape, original.dtype)

# akses per pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(brightness * (np.log(1 + original[y,x,c])), 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

4. Implementasikan transformasi grayscale menggunakan metode averaging, lightness,
dan luminance pada Google Colaboratory menggunakan formula yang terdapat pada
bagian Ulasan Teori, sehingga menghasilkan keluaran seperti berikut:
a. Averaging

```python
ktp = cv.imread('/content/drive/MyDrive/File PCVK/KTP1.jpeg')
ktp_avg = cv.cvtColor(ktp, cv.COLOR_BGR2RGB)
plt.imshow(ktp_avg)
```

```python
# Averaging
print('Transformasi Grayscale - Metode Averaging')
print('-----------------------------------------')
ktp_af = cv.imread('/content/drive/MyDrive/File PCVK/KTP1.jpeg')

# Konversi citra ke grayscale menggunakan metode averaging
gray_image = np.mean(ktp_af, axis=2).astype(np.uint8)

# Menampilkan citra asli dan citra grayscale
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Citra Asli')
plt.imshow(cv.cvtColor(ktp_af, cv.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Citra Grayscale (Averaging)')
plt.imshow(gray_image, cmap='gray')
plt.show()
```

b. Lightness

```python
# Lightness
print('Transformasi Grayscale - Metode Lightness')
print('-----------------------------------------')
ktp_li = cv.imread('/content/drive/MyDrive/File PCVK/KTP1.jpeg')

grayscale_image =  (np.max(ktp_li, axis = 2))/2 + (np.min(ktp_li, axis = 2))/2

# Menampilkan citra RGB dan citra grayscale
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.title('Citra RGB')
plt.imshow(cv.cvtColor(ktp_li, cv.COLOR_BGR2RGB))

plt.subplot(122)
plt.title('Citra Grayscale (Lightness)')
plt.imshow(grayscale_image, cmap='gray')
plt.show()
```

c. Luminance

```python
# Luminance
print('Transformasi Grayscale - Metode Luminance')
print('-----------------------------------------')
ktp_lu = cv.imread('/content/drive/MyDrive/File PCVK/KTP1.jpeg')

# Konversi citra ke mode warna Luv
luv_image = cv.cvtColor(ktp_lu, cv.COLOR_BGR2Luv)

# Ambil komponen L (Luminance)
luminance_image = luv_image[:, :, 0]

# Menampilkan citra asli dan citra grayscale dengan metode Luminance
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.title('Citra Asli')
plt.imshow(cv.cvtColor(ktp_lu, cv.COLOR_BGR2RGB))

plt.subplot(122)
plt.title('Citra Grayscale (Luminance)')
plt.imshow(luminance_image, cmap='gray')

plt.show()
```

5. Tampilkanlah warna tertentu pada citra, dan ubah warna lain menjadi grayscale. Misal,
tampilkan warna biru pada citra masukan dan ubah bagian lain yang tidak berwarna
biru menjadi grayscale seperti pada contoh berikut:

```python
# Metode gray-Scale pada KTP
img = cv.imread('/content/drive/MyDrive/File PCVK/KTP2.jpg')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
gray = cv.merge([gray, gray, gray])

hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

lower = np.array([55, 120, 120])
upper = np.array([185, 220, 223])

mask = cv.inRange(hsv, lower, upper)

colored_output = cv.bitwise_and(img, img, mask=mask)

gray_output = cv.bitwise_and(gray, gray, mask=255-mask)

result = cv.add(colored_output, gray_output)

plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.imshow(cv.cvtColor(colored_output, cv.COLOR_BGR2RGB))
plt.title('Colored Output')

plt.subplot(1, 3, 2)
plt.imshow(cv.cvtColor(gray_output, cv.COLOR_BGR2RGB))
plt.title('Gray Output')

plt.subplot(1, 3, 3)
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.title('Result')

plt.show()
```
